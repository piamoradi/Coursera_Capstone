Scrape table of postal code information from Wikipedia website

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
dfs = pd.read_html(url)
tor_codes = dfs[0]
tor_codes
```

Find subset of dataframe witout "Not assigned" boroughs

```python
tor_codes2 = tor_codes.loc[tor_codes['Borough']!='Not assigned']
tor_codes2
```

Reset index to create final version of Toronto postal code dataframe

```python
tor_codes3 = tor_codes2.reset_index()
tor_codes3.drop(['index'],axis=1)
```
